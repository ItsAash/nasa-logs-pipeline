```python
kafka_bootstrap_server = "0.tcp.in.ngrok.io:11325" 
topic_name = "nasa-logs"

# Define a unique path for your checkpoints
# This helps Spark remember where it left off in the Kafka stream
checkpoint_path = "/Volumes/workspace/default/logs_volume/test"

# Clean up previous checkpoint to avoid state errors
dbutils.fs.rm(checkpoint_path, True)

raw_stream_df = spark.readStream \
  .format("kafka") \
  .option("kafka.bootstrap.servers", kafka_bootstrap_server) \
  .option("subscribe", topic_name) \
  .option("startingOffsets", "earliest") \
  .load()

df = raw_stream_df.selectExpr("CAST(key AS STRING)", "CAST(value AS STRING)")

# 4. Display the Data
display(df, checkpointLocation = checkpoint_path)

```
