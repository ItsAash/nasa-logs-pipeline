```python
RAW_LOG_PATH = "/Volumes/workspace/default/logs_volume/NASA_access_log_Jul95.gz"
BRONZE_PATH = "/Volumes/workspace/default/logs_volume/bronze"
```

```python
# Read the raw log file as plain text (one row per log line)
raw_df = spark.read.text(RAW_LOG_PATH)
print(f"Total log lines: {raw_df.count():,}")
raw_df.show(5, truncate=False)
```

```python
# Write raw data as a Delta table - this is Bronze layer

raw_df.write.format("delta").mode("overwrite").save(BRONZE_PATH)
print("Bronze layer written successfully.")
```

```python
# Verifying
bronze_df = spark.read.format("delta").load(BRONZE_PATH)
print(f"Rows in Bronze Delta table: {bronze_df.count():,}")
```
